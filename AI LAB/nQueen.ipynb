```python
import random

N = int(input())

def random_state():
    return [random.randint(0, N-1) for _ in range(N)]

def attacks(state, row, col):
    cnt=0
    for i in range(col):
      if state[i] ==row or abs(state[i]-row)==abs(i-col):
        cnt+=1
    return cnt
    
def total_attacks(state):
    return sum(attacks(state, state[col], col) for col in range(N))

def hill_climbing():
    current_state = random_state()
    attack_cnt = total_attacks(current_state)
    
    attempt_cnt=1
    print("     Attempt : Total_attacks")
    while attack_cnt > 0:
        next_state = current_state
        col = random.randint(0, N-1)
        row = random.randint(0, N-1)
        next_state[col] = row

        print("\t",attempt_cnt ," : ", attack_cnt)
        #print_board(current_state)
        new_attack_cnt = total_attacks(next_state)

        if new_attack_cnt <= attack_cnt:
            current_state, attack_cnt = next_state, new_attack_cnt
        attempt_cnt+=1

        
    print("        ",attempt_cnt, " : ", attack_cnt)
    return current_state

def print_board(state):
  #print(state)
  print("Queen Positions : ")
  for row in state:
        for col in range(N):
            if row==col:
                print(1,end=" ")
            else:
                print(0,end=" ")
        print()
     

solution = hill_climbing()
print_board(solution)
```
