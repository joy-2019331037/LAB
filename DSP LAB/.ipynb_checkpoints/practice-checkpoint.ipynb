```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
Fs=210
f=10
L=50
s=np.zeros(L)
for n in range(L):
    s[n] = np.cos(2*np.pi*f*n/Fs)

plt.stem(s)
```

```python
plt.plot(s)
```

```python
d = np.zeros(L)
d[L//2]=1
x = np.arange(-L//2,L//2)
plt.stem(x,d)
print(x.shape)
print(d.shape)
```
