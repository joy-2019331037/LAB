```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
F=10
Fs=400
L=100

signal = np.zeros(L)

for n in range(L):
    signal[n] = np.sin(2*np.pi*(F/Fs)*n)


plt.plot(signal)
```

```python
import random
noise = np.zeros(L)

for n in range(L):
    noise[n] = random.random()

plt.plot(noise)
```

```python
noisy_signal = noise + signal

plt.plot(noisy_signal)
```

Moving Average Filter

```python
y_MAF=np.zeros(L)
for i in range(1,L-1):
    y_MAF[i] = (noisy_signal[i-1]+noisy_signal[i]+ noisy_signal[i+1])/3

plt.plot(y_MAF)
```

Accumulator

```python
y_ACC=np.zeros(L)

for i in range(L):
    y_ACC[i]=sum(y_MAF[:i+1])

plt.plot(y_ACC)
```

First order derivative

```python
y_FOD = np.zeros(L)

for i in range(1,L-1):
    y_FOD[i] = y_ACC[i] -y_ACC[i-1]

plt.plot(y_FOD)
```

is there any order or not??

Median

```python
y_mod = np.zeros(L)

for i in range(2,L-2):
    y_mod[i] = sorted([y_FOD[i-2],y_FOD[i-1],y_FOD[i],y_FOD[i+1],y_FOD[i+2]])[2]

plt.plot(y_mod)
```
