```python
import numpy as np 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("error")
```

```python
N=256
F1=50
F2=100
F3=150
Fs=1000

x = np.zeros(N)

for n in range(N):
    x[n] = np.cos(2*np.pi*(F1/Fs)*n) + np.cos(2*np.pi*(F2/Fs)*n) + np.cos(2*np.pi*(F3/Fs)*n)

plt.plot(x)
```

DFT (using formula)

```python
X = np.zeros(N, dtype='complex')

for k in range(N):
    for n in range(N):
        X[k]+= x[n] * np.exp(-1j*2*np.pi*k*n/N)

plt.plot(abs(X))
```

FFT (using library)

```python
fft = np.fft.fft(x)

plt.plot(abs(fft))
```

IDFT (using formula)

```python
idft = np.zeros(N,dtype='complex')

for n in range(N):
    for k in range(N):
        idft[n]+= X[k]*np.exp(1j*2*np.pi*k*n/N)

idft = idft/N

plt.plot(abs(idft))
```

IDFT (using library)

```python
idft_library = np.fft.ifft(X)
print(np.allclose(idft,idft_library))
```
