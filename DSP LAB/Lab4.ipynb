```python

import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.random.randint(0,9,5)
h= np.random.randint(-10,10,5)

print("x = {}, y = {}".format(x,h) )

y_builtIn = np.convolve(x,h)
N1 = np.size(x)
N2 = np.size(h)
N=N1+N2-1
y_manual = np.zeros(N)

m = N-N1
n= N-N2

x = np.pad(x,(0,m),'constant')
h = np.pad(h,(0,n),'constant')

for n in range(N):
    for k in range(N):
        if n>=k:
            y_manual[n]+=x[n-k]*h[k]

print("Built In Function result : ", y_builtIn)
print("Manual result : ", y_manual)

print("Do both have the same output? : ", np.allclose(y_builtIn,y_manual))



plt.figure(figsize=(10,4))
plt.subplot(131)
plt.title('x(n)')
plt.stem(x)
plt.grid(True)

plt.subplot(132)
plt.title('h(n)')
plt.stem(h)
plt.grid(True)

plt.subplot(133)
plt.title('convolutional')
plt.stem(y_manual)
plt.grid(True)

plt.grid(True)
```
