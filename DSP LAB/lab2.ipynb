```python
import numpy as np
import matplotlib.pyplot as plt
```

Continuous Time Signal

```python
t_cont = np.linspace(0,5,1000)

f = 2
A = 1

signal_cont = A * np.sin(2*np.pi*f*t_cont)

#making figure size 10:4 
plt.figure(figsize=(10,4))

#plotting the singnal
plt.plot(t_cont,signal_cont)
plt.xlabel('Time')
plt.ylabel('Amptlitude')
plt.title('Continuous Time Signal')


#setting x limit from 0 to last element of t
plt.xlim(0,t_cont[-1])
#setting ylimit (max value of signal(1), and min value of signal(-1))
plt.ylim(signal_cont.min(),signal_cont.max())

plt.grid(True)
```

Discrete Time Signal

```python
Fs = 1000
F = 10
L = 100
signal = np.zeros(L)

for n in range(L):
    signal[n] = np.sin(2*np.pi*(F/Fs)*n)

plt.ylim(signal.min(),signal.max())
plt.stem(signal) 
plt.title('Discrete Time Signal')
```

Unit Impulse Response

```python
L = 20
y=np.zeros(L)
y[L//2] = 1
x=np.arange(-L//2,L//2)

plt.title('Unit Impulse Signal')
plt.stem(x,y)
```

Unit Step Signal


```python
L = 50
y=np.zeros(L)
y[L//2:] = 1
x=np.arange(-L//2,L//2)

plt.title('Unit Impulse Signal')
plt.stem(x,y)
```

Unit Ramp Signal

```python
L=50
y = np.maximum(0,np.arange(-L//2,L//2))
x=np.arange(-L//2,L//2)

plt.stem(x,y)
plt.title('Unit Ramp Signal')
```

Exponential Signal

```python
L=100
x=np.zeros(L)

a=-0.9
for n in range(L):
    x[n] = a**n

plt.text(80,0.75,'-1<a<0',color='red')
plt.stem(x)
```
